## Ciência de dados
</div>

Com o Colab, você pode aproveitar todo o potencial das conhecidas bibliotecas Python para analisar e ver dados. A célula de códigos abaixo usa <strong>numpy</strong> para gerar dados aleatórios e <strong>matplotlib</strong> para visualizá-los. Para editar o código, basta clicar na célula e começar a editar.

```
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```

```
import PIL
import cv2
import matplotlib.pyplot as plt
import numpy as np

```

```
from PIL import Image
Image.open("eu.jpg")
```

```
teste=cv2.imread('/content/eu.jpg')
```

```
plt.imshow(teste[:,:,0], cmap="gray")
```

```
lampada = cv2.imread('lampada.jpg')
blur=cv2.blur(lampada,(5,5))
plt.imshow(lampada), plt.title('borrada')
plt.subplot(122),plt.imshow(blur),plt.title('Borrada')
plt.show()
```

```
imgGray = cv2.cvtColor(teste, cv2.COLOR_BGR2GRAY)
plt.imshow(imgGray)
```

```
imgHSV = cv2.cvtColor(teste, cv2.COLOR_BGR2HSV)
plt.imshow(imgHSV)
```

```
imgCanny = cv2.Canny(teste, 100, 100)
plt.imshow(imgCanny)
```

```
kernel = np.ones((5,5), np.uint8)
imgDialation = cv2.dilate(imgCanny, kernel, iterations=1)
plt.imshow(imgDialation)

```

```
kernel = np.ones((5, 5), np.uint8)
imgEroded = cv2.erode(imgDialation, kernel, iterations=1)
plt.imshow(imgEroded)
```

```
imgResize1 = cv2.resize(teste, (820, 1000))
plt.imshow(imgResize1)
```
